```python
import pandas as pd
import numpy as np
import joblib
from sklearn.preprocessing import StandardScaler, RobustScaler
```

```python
raw_data = pd.read_csv('data/Customer_Data.csv', sep=',')
raw_data
```

```python
raw_data.shape
```

```python
raw_data.describe()
```

```python
raw_data.columns
```

```python
raw_data.drop(columns=['CUST_ID'], axis=1, inplace=True)
```

```python
raw_data.columns
```

```python
print(f"Duplicates: {raw_data.duplicated().sum()}")
```

```python
print(f"Nulls: {raw_data.isna().sum()}")
```

```python
raw_data.dropna(inplace=True)
```

```python
q1 = raw_data.quantile(0.25)
q3 = raw_data.quantile(0.75)
iqr = q3 - q1
outliers = (raw_data < (q1 - 1.5 * iqr)) | (raw_data > (q3 + 1.5 * iqr))
total_outliers = outliers.sum()
print(f"Total outliers: {total_outliers}")
```

```python
print(f"Number of outliers: {outliers.any(axis=1).sum()}")
```

```python
robust_scaler = RobustScaler()
scaled_data = robust_scaler.fit_transform(raw_data)
scaled_df = pd.DataFrame(scaled_data, columns=raw_data.columns)
scaled_df
```

```python
scaled_df.to_csv('data/Customer_Data_Processed.csv', index=False)
```

```python
joblib.dump(robust_scaler, 'obj/robust_scaler.joblib')
```
